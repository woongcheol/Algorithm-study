# Programmers 프린터 문제

### 문제 설명
트럭 여러 대가 강을 가로지르는 일차선 다리를 정해진 순으로 건너려 합니다. 모든 트럭이 다리를 건너려면 최소 몇 초가 걸리는지 알아내야 합니다. 다리에는 트럭이 최대 bridge_length대 올라갈 수 있으며, 다리는 weight 이하까지의 무게를 견딜 수 있습니다. 단, 다리에 완전히 오르지 않은 트럭의 무게는 무시합니다.

예를 들어, 트럭 2대가 올라갈 수 있고 무게를 10kg까지 견디는 다리가 있습니다. 무게가 [7, 4, 5, 6]kg인 트럭이 순서대로 최단 시간 안에 다리를 건너려면 다음과 같이 건너야 합니다

따라서, 모든 트럭이 다리를 지나려면 최소 8초가 걸립니다.

solution 함수의 매개변수로 다리에 올라갈 수 있는 트럭 수 bridge_length, 다리가 견딜 수 있는 무게 weight, 트럭 별 무게 truck_weights가 주어집니다. 이때 모든 트럭이 다리를 건너려면 최소 몇 초가 걸리는지 return 하도록 solution 함수를 완성하세요.

### 제한 조건
bridge_length는 1 이상 10,000 이하입니다.
weight는 1 이상 10,000 이하입니다.
truck_weights의 길이는 1 이상 10,000 이하입니다.
모든 트럭의 무게는 1 이상 weight 이하입니다

```python
# 첫번째 시도
import queue
def solution(bridge_length, weight, truck_weights):
    answer = 0
    sum_weight = 0
    cnt = 0
    exit = [[]]
    bridge_max = queue.Queue()
    
    # 트럭이 모두 건널 때까지 반복문 실행
    while truck_weights:
        for i in range(len(truck_weights)):
            # 소요 시간 관리
            cnt += 1
            

            # 트럭 이동 가능여부 확인, 상태 등록
            if sum_weight + truck_weights[i] <= weight and bridge_max.qsize() <= bridge_length:
                sum_weight += truck_weights[i]
                bridge_length -= 1
                bridge_max.put(truck_weights[i])

            for j in range(len(exit)):
                if exit[j]:
                    exit.append([cnt])

                exit[j].append(cnt)

            # 이동 시간 관리 - 트럭 도착시 이동 상태 삭제
            if len(exit[0]) > bridge_length:
                bridge_max.get()
                del truck_weights[0]

    answer = cnt[-1][-1]
    return answer
```

```python
# 2차 시도
def solution(priorities, location):
    answer = 0
    cnt = 0
    stop = 0

    # 출력물 인덱스 저장
    a = [number for number in range(len(priorities))]

    # 중요도 탐색
    while True:

        # 중요도 사전 탐색
        for i in range(1, len(priorities)):
            if priorities[0] < priorities[i]:
                stop = 1

        # 중요도 탐색 순환 - location
        if a[0] == location and stop == 1:
            priorities.append(priorities[0])
            del priorities[0]
            a.append(a[0])
            del a[0]
            stop = 0

        # 중요도 탐색 순환 - general
        elif stop == 1:
            priorities.append(priorities[0])
            del priorities[0]
            a.append(a[0])
            del a[0]
            stop = 0      

        # location 출력시
        elif a[0] == location and stop == 0:
            cnt += 1
            answer = cnt
            break

        # general 출력시
        elif stop == 0:
            del priorities[0]
            del a[0]
            cnt += 1

    return answer

priorities = [2, 1, 3, 2]
location = 2
print(solution(priorities, location))
```
